```python
!pip install openai
from openai import OpenAI
import json
from datetime import datetime

# ðŸ”‘ paste your API key here safely (temporary, not uploaded)
client = OpenAI(api_key="YOUR_API_KEY_HERE")

logs = []

def office_bot(msg):
    response = client.responses.create(
        model="gpt-4o-mini",
        input=f"You are a polite UK Smart Office assistant. {msg}"
    )
    reply = response.output[0].content[0].text
    logs.append({
        "time": datetime.now().isoformat(),
        "message": msg,
        "reply": reply
    })
    print(reply)

# â–¶ Example test
office_bot("Summarize today's meeting in 2 bullet points.")
```

```python
with open("logs.json", "w") as f:
    json.dump(logs, f, indent=2)
print("âœ… Saved logs.json")
```
