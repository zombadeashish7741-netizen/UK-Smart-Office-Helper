```python
from datetime import datetime
from openai import OpenAI

client = OpenAI(api_key="client = OpenAI(api_key="YOUR_API_KEY_HERE")
")
logs = []

def smart_helper(msg, region="UK"):
    if region != "UK":
        return {"error": "Only UK region supported"}

    # Logic for basic cases
    text = msg.lower()
    if "course" in text or "join" in text:
        reply = "Thanks for asking about the course! We’ll contact you soon."
    elif "complaint" in text:
        reply = "We got your complaint. We’ll fix it right away."
    else:
        # Real AI fallback
        r = client.responses.create(
            model="gpt-4o-mini",
            input=f"Reply like a polite UK Smart Office assistant: {msg}"
        )
        reply = r.output[0].content[0].text

    log_item = {
        "time": datetime.utcnow().isoformat(),
        "msg": msg,
        "reply": reply
    }
    logs.append(log_item)
    return reply
```

```python
print(smart_helper("I want to join your AI course"))
print(smart_helper("I have a complaint about my laptop"))
print(logs)
```

```python
import json

# Save logs to a file
with open("logs.json", "w") as f:
    json.dump(logs, f, indent=2)

print("Saved", len(logs), "entries to logs.json")

```
